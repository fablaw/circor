```python
import numpy as np
import pandas as pd
import librosa
import librosa.display
```

```python
file_1='../raw_data/1.0.3/audio_raw/14241_PV.wav'
x_1, sr_1=librosa.load(file_1, sr=None)
len(x_1), sr_1
```

```python
file_2='../raw_data/1.0.3/audio_treated/14241_PV.wav'
x_2, sr_2=librosa.load(file_1, sr=None)
len(x_2), sr_2
```

```python
y=[]
for i in range(0, len(x_2)-1):
    y.append(abs(x_2[i+1]-x_2[i]))
y
```

```python
import matplotlib.pyplot as plt
plt.plot(y)
```

```python
from scipy import signal
from scipy.ndimage import gaussian_filter1d
sigma=np.std(y)

dataFiltered = gaussian_filter1d(y, sigma=sigma)
tMax = signal.argrelmax(dataFiltered)[0]

tMax
```
