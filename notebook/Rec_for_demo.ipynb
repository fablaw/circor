```python

D = librosa.stft(x_1[0:50000], n_fft=446, hop_length=224)
S_db = librosa.amplitude_to_db(np.abs(D), ref=np.max)

librosa.display.specshow(S_db, y_axis="log", sr=sr, hop_length=1024, x_axis="time")
```
